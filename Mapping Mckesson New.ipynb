```python
#import file which has the weblinks for each product
Mckesson_PM_data=pd.read_excel("Mckesson Second_collated_done.xlsx")
```

```python
Mckesson_PM_data=Mckesson_PM_data[Mckesson_PM_data['WebLink'].notnull()]
```

```python
#output dataframe to store the scrapped the results
crawled_op=pd.DataFrame()
```

```python
#loop through each row of the input file and get the url link to product
for idx,row in Mckesson_PM_data.iterrows():
    URL=str(row['WebLink']).strip()
#     print("SPN to search:  {}".format(spn_file))
#     print(URL)
    
    try:
        r=requests.get(URL)
    except:
        print("Not processed !!!!!!!!!!!!!! {}".format(URL))
        continue
        
    if (r.status_code==200):
        soup = BeautifulSoup(r.text, 'html.parser')
        tbl=soup.select_one('table') # In mckesson there is only one table which has details hence look for a table
        if tbl:
            list_of_td=tbl.find_all('td')
            to_append={} #declare a dict to store the scrapped values
            length=len(list_of_td)
            #Loop through each row in table and get the Key(Header) and value and add it to dict
            for i in range(0,length,2):
                to_append[list_of_td[i].contents.pop()]=list_of_td[i+1].contents.pop()
            to_append['Secondary_Key']=str(row['Secondary_Key'])
            to_append['is_sourced']="True"
            crawled_op=crawled_op.append(to_append,ignore_index=True)
            print(URL)
            print(len(crawled_op))

crawled_op.to_csv('Url_sourced_mckesson_data_second_collated.csv')
```
